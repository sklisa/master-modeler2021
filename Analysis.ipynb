```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('SentimentAnalysis.csv')
df['key'] = df.apply(lambda row: int(row['filename'][:-4]), axis=1)
df.drop(columns=['filename'], axis=1, inplace=True)
```

```python
dataset = pd.read_csv('dataset_0320.csv')
df = df.join(dataset.set_index('key'), on='key')
```

```python
df
```

```python
ax = sns.regplot(x="Positiv_GI", y="engagement_rate_label", data=df)
ax.set(ylim=(0, 2))
```

```python
ax = sns.boxplot(x="engagement_rate_label", y="certainty_component", data=df)
ax.set(ylim=(0, 1))
```

```python
ax = sns.catplot(y="evening", hue="engagement_rate_label", kind='count', data=df)
```

```python
ax = sns.catplot(y="name_num", x="engagement_rate_label", kind='bar', data=df)
```
