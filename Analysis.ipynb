```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('SentimentAnalysis.csv')
df['key'] = df.apply(lambda row: int(row['filename'][:-4]), axis=1)
df.drop(columns=['filename'], axis=1, inplace=True)
```

```python
dataset = pd.read_csv('dataset_0320.csv')
df = df.join(dataset.set_index('key'), on='key')
```

```python
ax = sns.regplot(x="Positiv_GI", y="engagement_rate_label3", data=df)
ax.set(ylim=(0, 2))
```

```python
ax = sns.boxplot(x="weighted_engagement_label3", y="positive_adjectives_component", data=df)   
# positive_adjectives_component; certainty_component; affect_friends_and_family_component
# ax.set(ylim=(0, 1))
```
